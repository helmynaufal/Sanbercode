<h2>Soal 1:</h2>

Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# code here
# exsplisit

import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(x,cos, c='r', label="Cosx")
ax.plot(x,cos3, c='g', label="3 Cosx", ls="--")
ax.plot(x,sin, c='b', label="Sin(x-90)", ls="-.")
ax.plot(x,sin2, c='y', label="2 Sinx", ls="None", marker='x')
ax.set_title("Grafik Cos Sin")
ax.set_xlabel("X-Axis Label")
ax.set_ylabel("Y-Axis Label")
plt.legend(loc="center")
plt.show()
```

```python
# implisit
plt.figure(figsize=(12,8))
plt.plot(x,cos, c='r', label="Cosx")
plt.plot(x,cos3, c='g', label="3 Cosx", ls="--")
plt.plot(x,sin, c='b', label="Sin(x-90)", ls="-.")
plt.plot(x,sin2, c='y', label="2 Sinx", ls="None", marker='x')
plt.xlabel("X-Axis Label")
plt.ylabel("Y-Axis Label")
plt.title("Grafik Cos Sin")
plt.legend(loc="center")
plt.show()
```
